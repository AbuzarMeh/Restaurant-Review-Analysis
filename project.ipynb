```python
from transformers import pipeline
# Load sentiment analysis pipeline
def sentiment_analysis_pipeline():
    return pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

# Sentiment analysis function
def analyze_sentiment(system_prompt, user_prompt):
    """
    Analyzes sentiment based on a system prompt and user prompt.
    
    Args:
        system_prompt (str): Context or background (not directly used in analysis but can be logged or extended).
        user_prompt (str): The text to analyze for sentiment.
    
    Returns:
        dict: Sentiment result including label (e.g., Positive, Negative) and confidence score.
    """
    # Initialize pipeline
    sentiment_pipeline = sentiment_analysis_pipeline()
    
    # Analyze user prompt sentiment
    sentiment_result = sentiment_pipeline(user_prompt)
    
    # Return analysis result
    return {
        "system_prompt": system_prompt,
        "user_prompt": user_prompt,
        "sentiment": sentiment_result[0]  # Includes 'label' and 'score'
    }

# Example usage
# if __name__ == "_main_":
system_prompt = "Analyze the sentiment of the user's review."
user_prompt = "The food was amazing, but the service was a bit slow."

result = analyze_sentiment(system_prompt, user_prompt)
print("Printing analysis")
print(result)
```

```python
import anthropic

client = anthropic.Anthropic()
review = ""
prompt = f"""
        You are an assistant trained to analyze customer reviews. Your task is to:
        1. Extract and categorize any comments related to the quality of food.
        2. Extract and categorize any comments related to staff/service.
        3. Do not include any personal information of the reviewers.
        4. Ensure no hallucinations: Do not add any data irrelevant to the context.
        Here is the review: "{review}"

        Respond in JSON format with the following structure:
        {{
            "food_comments": ["List of comments about food"],
            "staff_comments": ["List of comments about staff/service"]
        }}
        """
message = client.messages.create(
    
    model="claude-3-5-sonnet-20241022",
    max_tokens=1000,
    temperature=0,
    system="You are an assistant trained to analyze and categorize customer reviews.",
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ]
)
print(message.content)
```
